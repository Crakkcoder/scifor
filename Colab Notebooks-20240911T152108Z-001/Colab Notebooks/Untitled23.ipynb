```python
import tkinter as tk
from tkinter import ttk

def submit_form():
    name = name_entry.get()
    age = age_entry.get()
    gender = gender_var.get()
    address = address_text.get("1.0", tk.END).strip()
    email = email_entry.get()
    contact = contact_entry.get()
    country = country_entry.get()
    state = state_entry.get()
    diseases = []
    if cold_var.get():
        diseases.append("cold")
    if cough_var.get():
        diseases.append("cough")
    if fever_var.get():
        diseases.append("fever")
    if headache_var.get():
        diseases.append("headache")

    # Print form data (for demonstration)
    print("Name:", name)
    print("Age:", age)
    print("Gender:", gender)
    print("Address:", address)
    print("Email:", email)
    print("Contact No:", contact)
    print("Country:", country)
    print("State:", state)
    print("Diseases:", ", ".join(diseases))

# Create the main window
root = tk.Tk()
root.title("COVID Vaccine Registration Form")

# Create and place labels and entry fields
tk.Label(root, text="Name of the Visitor").grid(row=0, column=0, padx=10, pady=5, sticky="w")
name_entry = tk.Entry(root)
name_entry.grid(row=0, column=1, padx=10, pady=5, sticky="w")

tk.Label(root, text="Age of the Visitor").grid(row=1, column=0, padx=10, pady=5, sticky="w")
age_entry = tk.Entry(root)
age_entry.grid(row=1, column=1, padx=10, pady=5, sticky="w")

tk.Label(root, text="Gender").grid(row=2, column=0, padx=10, pady=5, sticky="w")
gender_var = tk.StringVar()
tk.Radiobutton(root, text="Male", variable=gender_var, value="Male").grid(row=2, column=1, padx=10, pady=5, sticky="w")
tk.Radiobutton(root, text="Female", variable=gender_var, value="Female").grid(row=2, column=2, padx=10, pady=5, sticky="w")

tk.Label(root, text="Address").grid(row=3, column=0, padx=10, pady=5, sticky="w")
address_text = tk.Text(root, height=4, width=30)
address_text.grid(row=3, column=1, padx=10, pady=5, sticky="w")

tk.Label(root, text="Email ID").grid(row=4, column=0, padx=10, pady=5, sticky="w")
email_entry = tk.Entry(root)
email_entry.grid(row=4, column=1, padx=10, pady=5, sticky="w")

tk.Label(root, text="Contact No").grid(row=5, column=0, padx=10, pady=5, sticky="w")
contact_entry = tk.Entry(root)
contact_entry.grid(row=5, column=1, padx=10, pady=5, sticky="w")

tk.Label(root, text="Country").grid(row=6, column=0, padx=10, pady=5, sticky="w")
country_entry = tk.Entry(root)
country_entry.grid(row=6, column=1, padx=10, pady=5, sticky="w")

tk.Label(root, text="State").grid(row=7, column=0, padx=10, pady=5, sticky="w")
state_entry = tk.Entry(root)
state_entry.grid(row=7, column=1, padx=10, pady=5, sticky="w")

tk.Label(root, text="Do you have any of the following diseases?").grid(row=8, column=0, padx=10, pady=5, sticky="w")

cold_var = tk.BooleanVar()
cough_var = tk.BooleanVar()
fever_var = tk.BooleanVar()
headache_var = tk.BooleanVar()

tk.Checkbutton(root, text="Cold", variable=cold_var).grid(row=8, column=1, padx=10, pady=5, sticky="w")
tk.Checkbutton(root, text="Cough", variable=cough_var).grid(row=8, column=2, padx=10, pady=5, sticky="w")
tk.Checkbutton(root, text="Fever", variable=fever_var).grid(row=8, column=3, padx=10, pady=5, sticky="w")
tk.Checkbutton(root, text="Headache", variable=headache_var).grid(row=8, column=4, padx=10, pady=5, sticky="w")

# Submit button
submit_button = tk.Button(root, text="Submit", command=submit_form)
submit_button.grid(row=9, column=0, columnspan=5, padx=10, pady=10)

# Run the application
root.mainloop()
```
