```python
# function to add two numbers
def add_numbers(a,b):
    return a + b

# function to subtract two numbers
def subtract_numbers(a,b):
    return a - b

# function to multiply two numbers
def multiply_numbers(a,b):
    return a * b

a = 20
b = 10

print(f"{a} + {b} = {add_numbers(a,b)}")
print(f"{a} - {b} = {subtract_numbers(a,b)}")
print(f"{a} * {b} = {multiply_numbers(a,b)}")

```

```python
# Function to calculate simple interest
def calculate_simple_interest(principal, rate, time):
    return (principal * rate * time) / 100

# Function to calculate compound interest
def calculate_compound_interest(principal, rate, time):
    return principal * (pow((1 + rate / 100), time)) - principal

# Example usage:
principal_amount = 1000
interest_rate = 5
time_period = 2

simple_interest = calculate_simple_interest(principal_amount, interest_rate, time_period)
compound_interest = calculate_compound_interest(principal_amount, interest_rate, time_period)

print(f"Simple Interest: {simple_interest}")
print(f"Compound Interest: {compound_interest}")
```

```python
# Function to check if a number is a perfect square
def is_perfect_square(num):
    if num < 0:
        return False
    if num == 0:
        return True

    x = num // 2
    seen = set([x])

    while x * x != num:
        x = (x + (num // x)) // 2
        if x in seen:
            return False
        seen.add(x)

    return True

# Example usage:
number = 25
if is_perfect_square(number):
    print(f"{number} is a perfect square.")
else:
    print(f"{number} is not a perfect square.")
```

```python
# Function to convert a number into words with hyphens
def toWord(num):
    words = {
        '0': 'zero', '1': 'one', '2': 'two', '3': 'three', '4': 'four',
        '5': 'five', '6': 'six', '7': 'seven', '8': 'eight', '9': 'nine'
    }

    num_str = str(num)
    result = ""

    for digit in num_str:
        result += words[digit] + "-"

    # Remove the last hyphen
    result = result[:-1]

    return result

# Example usage:
number1 = 89
number2 = 398

print(f"{number1} => {toWord(number1)}")
print(f"{number2} => {toWord(number2)}")
```
