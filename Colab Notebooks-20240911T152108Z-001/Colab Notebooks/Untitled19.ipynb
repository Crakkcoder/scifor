```python
def euler_method(func,initial_x,initial_y,step_size,num_steps):
   x = initial_x
   y = initial_y
   for i in range(num_steps):
       y = y + step_size * func(x,y)
       x = x + step_size
       return x,y

# Example usage:
def dy_dx(x,y):
    return x + y

x_final,y_final = euler_method(dy_dx,initial_x=0,initial_y=1,step_size=0.1,num_steps=10)

print(f"At x = {x_final},y = {y_final}")

```

```python
def runge_kutta(func,initial_x,initial_y,step_size,num_steps):
   x = initial_x
   y = initial_y
   for i in range(num_steps):
       k1 = step_size * func(x,y)
       k2 = step_size * func(x + step_size/2,y + k1/2)
       k3 = step_size * func(x + step_size/2,y + k2/2)
       k4 = step_size * func(x + step_size,y + k3)
       y = y + (k1 + 2*k2 + 2*k3 + k4)/6
       x = x + step_size
       return x,y

# Example usage:
def dy_dx(x,y):
    return x + y

x_final,y_final = runge_kutta(dy_dx,initial_x=0,initial_y=1,step_size=0.1,num_steps=10)

print(f"At x = {x_final},y = {y_final}")
```

```python
def secant_method(func,x0,x1,tolerance=1e-10,max_iterations=100):
    for i in range(max_iterations):
        fx1 = func(x1)
        if abs(fx1) < tolerance:
            return x1
        fx0 = func(x0)
        denominator = (fx1 - fx0)/(x1 - x0)
        x_next = x1 - fx1 / denominator
        x0,x1 = x1,x_next
    raise ValueError("Failed to coverage")

# Example usage:
def g(x):
    return x**3 - 2*x**2 - 5

root = secant_method(g,x0=0,x1=1)
print(f"Root found at x = {root}")
```

```python
def bisection_method(func,a,b,tolerance=1e-10,max_iterations=100):
    if func(a) * func(b) >= 0:
        raise ValueError("Function does not change sign over interval")

    for i in range(max_iterations):
        c = (a + b)/2
        if abs(func(c)) < tolerance:
            return c
        if func(c) * func(a) < 0:
            b = c
        else:
            a = c
        raise ValueError("Failed to coverage")



# Example usage:
def f(x):
    return x**3 - 2*x**2 - 5

root = bisection_method(f,a=0,b=1)
print(f"Root found at x = {root}")
```
