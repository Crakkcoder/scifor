```python
def count_desks(rows, columns):
  """
  This function takes the number of rows and columns as input and returns the total number of desks in the classroom.

  Args:
      rows: The number of rows of desks in the classroom.
      columns: The number of columns of desks in the classroom.

  Returns:
      The total number of desks in the classroom.
  """

  total_desks = rows * columns
  return total_desks

# Example usage
number_of_rows = int(input("Enter the number of rows of desks: "))
number_of_columns = int(input("Enter the number of columns of desks: "))
total_desks = count_desks(number_of_rows, number_of_columns)

print(f"The total number of desks in the classroom is {total_desks}.")



```

```python
def word_count(essay):
  """Counts the number of words in an essay.

  Args:
      essay: The essay text as a string.

  Returns:
      The number of words in the essay.
  """
  # Split the essay into a list of words
  words = essay.split()

  # Count the number of words
  word_count = len(words)

  return word_count

essay = input("Enter the essay text: ")

# Count the number of words in the essay
word_count = word_count(essay)

# Print the word count
print(f"The essay has {word_count} words.")

# Check if the essay meets the minimum word requirement
if word_count >= 250:
  print("The essay meets the minimum word requirement.")
else:
  print("The essay does not meet the minimum word requirement.")
```
