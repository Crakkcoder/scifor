```python
import requests

# Example URL for a GET request
url = 'https://api.quotable.io/random'

# Making a GET request
response = requests.get(url)

# Checking status
if response.status_code == 200:  # 200 means success
    print('GET request successful!')
    data = response.json()  # Assuming the response is JSON
    print('Quote:', data['content'])
else:
    print('GET request failed. Status code:', response.status_code)

```

```python
import requests

# Example URL for a POST request
url = 'https://example.com/api/post_endpoint'

# Example payload for POST request (data to be sent)
payload = {'key1': 'value1', 'key2': 'value2'}

# Making a POST request
response = requests.post(url, data=payload)

# Checking status
if response.status_code == 200:  # 200 means success
    print('POST request successful!')
else:
    print('POST request failed. Status code:', response.status_code)
```

```python
def get_random_quote():
    import requests

    url = 'https://api.quotable.io/random'

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad responses
        quote_data = response.json()
        return quote_data['content']
    except requests.exceptions.RequestException as e:
        print('Error fetching quote:', e)
        return None

# Example usage:
random_quote = get_random_quote()
if random_quote:
    print('Random Quote:', random_quote)
```

```python
def get_chuck_norris_joke():
    import requests

    url = 'https://api.chucknorris.io/jokes/random'

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad responses
        joke_data = response.json()
        return joke_data['value']
    except requests.exceptions.RequestException as e:
        print('Error fetching Chuck Norris joke:', e)
        return None

# Example usage:
chuck_norris_joke = get_chuck_norris_joke()
if chuck_norris_joke:
    print('Chuck Norris Joke:', chuck_norris_joke)
```
