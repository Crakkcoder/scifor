```python
path = "/presidents.csv"
df = pd.read_csv(path)
df.head()
```

```python
def change_date(date):
    return date.split('[')[0]
```

```python
df['Born'] = df['Born'].apply(change_date)
df['Died'] = df['Died'].apply(change_date)
```

```python
from datetime import datetime
df['Born'] = pd.to_datetime(df['Born'])
df.tail()
```
