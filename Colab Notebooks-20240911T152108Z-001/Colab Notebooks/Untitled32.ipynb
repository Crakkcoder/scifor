```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/amazon product.csv')

# Display the first few rows of the Dataframe
print(df.head())

# Example: Remove rows where the value in the 'Price' column is NaN (not a number)
df = df[df['Price'].notna()]

# Example: Remove a specific value from a column, e.g., remove rows where 'Category' is 'Electronics'
df = df[df['Category'] != 'Electronics']

# Save the modified DataFrame back to a CSV file
df.to_csv('modified_amazon_product.csv', index=False)
```
