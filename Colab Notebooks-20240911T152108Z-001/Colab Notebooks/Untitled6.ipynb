```python
from decimal import Decimal,getcontext

getcontext().prec = 6
a = Decimal('0.1')
b = Decimal('0.2')
c = a + b
print(c)
```

```python
a = 10 ** 100
print(a)
```

```python
a = 0.1 + 0.2
print(a)
```

```python
from fractions import Fraction
a = Fraction(1,3)
b = Fraction(2,3)
c = a + b
print(c)
```

```python
print(abs(-5))
print(round(3.14159,2))
print(pow(2,3))
```

```python
a = 3 + 4j
print(a.real)
print(a.imag)
```

```python
def sum_all(*args):
    """Sum all given numbers"""
    return sum(args)

total = sum_all(1,2,3,4,5)
print(total)
```

```python
def greet(name,greeting='Hello'):
    return '{},{}'.format(greeting,name)


print(greet('John'))
print(greet('John ,Hi'))
```

```python
def add(a,b):
    return a + b

result = add(4,6)
print(result)
```

```python
def multiply(a,b):
    return a * b

result = multiply(8,12)
print(result)
```

```python
def print_info(**kwargs):
    print(kwargs)

print_info(name = 'John',age = 50,city = 'Haripuram')
```

```python
def divide(a,b):
    return a / b


result = divide(16,24)
print(result)
```

```python
def square(a):
    return a ** 2

result = square(4)
print(result)
```

```python
square = lambda x: x ** 2
result = square(4)
print(result)
```
