```python
import tkinter as tk
from tkinter import messagebox
import random

class NamePickerApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Random Name Picker")

        # Initialize the list of student names
        self.student_names = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
        self.completed_names = []

        # Create and place widgets
        self.setup_widgets()

    def setup_widgets(self):
        # Frame for the random name display
        self.random_name_frame = tk.Frame(self.root)
        self.random_name_frame.pack(pady=10)

        self.random_name_label = tk.Label(self.random_name_frame, text="Random Name:", font=("Arial", 14))
        self.random_name_label.pack()

        self.random_name_display = tk.Label(self.random_name_frame, text="", font=("Arial", 24))
        self.random_name_display.pack()

        # Frame for completed names
        self.completed_names_frame = tk.Frame(self.root)
        self.completed_names_frame.pack(pady=10)

        self.completed_names_label = tk.Label(self.completed_names_frame, text="Completed Names:", font=("Arial", 14))
        self.completed_names_label.pack()

        self.completed_names_display = tk.Label(self.completed_names_frame, text="", font=("Arial", 14))
        self.completed_names_display.pack()

        # Button to pick a random name
        self.pick_button = tk.Button(self.root, text="Pick a Random Name", command=self.pick_random_name, font=("Arial", 14))
        self.pick_button.pack(pady=10)

    def pick_random_name(self):
        if not self.student_names:
            messagebox.showinfo("Info", "No more names left to pick!")
            return

        # Pick a random name from the list
        picked_name = random.choice(self.student_names)
        self.student_names.remove(picked_name)
        self.completed_names.append(picked_name)

        # Update the displayed name and completed names list
        self.random_name_display.config(text=picked_name)
        self.update_completed_names_display()

    def update_completed_names_display(self):
        completed_names_text = ", ".join(self.completed_names)
        self.completed_names_display.config(text=completed_names_text)

# Main execution
if __name__ == "__main__":
    root = tk.Tk()
    app = NamePickerApp(root)
    root.mainloop()
```
