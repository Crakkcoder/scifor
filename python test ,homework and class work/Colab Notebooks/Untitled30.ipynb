```python
import tkinter as tk
from tkinter import messagebox
import json

# Load questions from the JSON file
def load_questions(filename='questions.json'):
    with open(filename, 'r') as file:
        return json.load(file)

# Update the question and options in the GUI
def update_question():
    global current_question_index
    question = questions[current_question_index]

    question_label.config(text=question['question'])
    for i, option in enumerate(question['options']):
        option_vars[i].set(option)

    # Reset the selected answer
    selected_option.set(None)

# Check the selected answer and update the score
def check_answer():
    global current_question_index, score
    selected = selected_option.get()

    if selected == questions[current_question_index]['answer']:
        score += 1

    # Move to the next question
    current_question_index += 1

    if current_question_index < len(questions):
        update_question()
    else:
        show_result()

import tkinter as tk
from tkinter import messagebox
import json
from abc import ABC, abstractmethod

# Abstract class for a generic Quiz
class Quiz(ABC):
    def __init__(self, questions):
        self.questions = questions
        self.current_question_index = 0
        self.score = 0

    @abstractmethod
    def update_question(self):
        pass

    @abstractmethod
    def check_answer(self):
        pass

    @abstractmethod
    def show_result(self):
        pass

# Concrete class for a Tkinter-based Quiz
class TkinterQuiz(Quiz):
    def __init__(self, questions):
        super().__init__(questions)
        self.root = tk.Tk()
        self.root.title("Quiz Application")

        # Initialize GUI components
        self.question_label = tk.Label(self.root, text="", wraplength=400)
        self.question_label.pack(pady=10)

        self.option_vars = [tk.StringVar() for _ in range(4)]
        self.radio_buttons = [tk.Radiobutton(self.root, textvariable=var, variable=tk.StringVar(), value=var.get()) for var in self.option_vars]
        for rb in self.radio_buttons:
            rb.pack(anchor='w')

        self.selected_option = tk.StringVar()
        self.next_button = tk.Button(self.root, text="Next", command=self.check_answer)
        self.next_button.pack(pady=10)

        self.quit_button = tk.Button(self.root, text="Quit", command=self.show_result)
        self.quit_button.pack(pady=10)

        # Start with the first question
        self.update_question()

    def update_question(self):
        if self.current_question_index < len(self.questions):
            question = self.questions[self.current_question_index]
            self.question_label.config(text=question['question'])
            for i, option in enumerate(question['options']):
                self.option_vars[i].set(option)
            self.selected_option.set(None)
        else:
            self.show_result()

    def check_answer(self):
        selected = self.selected_option.get()
        correct_answer = self.questions[self.current_question_index]['answer']

        if selected == correct_answer:
            self.score += 1

        self.current_question_index += 1
        self.update_question()

    def show_result(self):
        messagebox.showinfo("Quiz Result", f"Your score is {self.score} out of {len(self.questions)}.")
        self.root.destroy()

    def run(self):
        self.root.mainloop()

# Function to load questions from a JSON file
def load_questions(filename='questions.json'):
    with open(filename, 'r') as file:
        return json.load(file)

# Main execution
if __name__ == "__main__":
    questions = load_questions()
    quiz = TkinterQuiz(questions)
    quiz.run()


```
