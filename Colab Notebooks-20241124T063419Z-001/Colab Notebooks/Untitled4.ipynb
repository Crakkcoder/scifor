```python
# Initialize an empty list to store student names
student_names = []

# Get the number of students
num_students = int(input("Enter the number of students in the class: "))

# Collect the names of students
print("Please enter the names of the students:")
for i in range(num_students):
    name = input(f"Name of student {i + 1}: ")
    student_names.append(name)

# Sort the list of student names in alphabetical order
student_names.sort()

# Print the sorted list with roll numbers
print("\nList of students in alphabetical order with roll numbers:")
for index, name in enumerate(student_names):
    # Roll number is index + 1
    roll_number = index + 1
    print(f"Roll Number {roll_number}: {name}")

# Optional: To show the final sorted list
print("\nFinal sorted list of student names:")
print(student_names)
```

```python
# Function to get the list of eligible football players
def get_eligible_players():
    players = []  # Initialize an empty list to store the players and their heights

    # Get the number of players
    num_players = int(input("Enter the number of players: "))

    # Collect the player names and heights
    for _ in range(num_players):
        name = input("Enter the player's name: ").strip()
        height = int(input("Enter the player's height in cm: "))

        if height < 167:
            print(f"{name} is not eligible to participate.")
            continue  # Skip to the next iteration of the loop if height is less than 167

        # Add eligible player to the list
        players.append((name, height))

    return players

# Main function
def main():
    eligible_players = get_eligible_players()

    if not eligible_players:
        print("No players are eligible to participate.")
    else:
        print("\nList of players eligible for the football competition:")
        for player in eligible_players:
            print(f"Name: {player[0]}, Height: {player[1]} cm")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
# Create an empty list for picnic items
picnic_items = []

# Number of items to be added
num_items = int(input("How many items do you want to add to the picnic list? "))

# Use a for loop to add items to the list
for i in range(num_items):
    item = input(f"Enter item {i + 1}: ")
    picnic_items.append(item)

# Print the final list of picnic items
print("\nFinal list of picnic items:")
for item in picnic_items:
    print(f"-{item}")


```
