```python
# base class:Rectangle
class Rectangle:
    def __init__(self):
        self.length = 0
        self.breadth = 0

    def setLength(self, length):
        self.length = length

    def setBreadth(self, breadth):
        self.breadth = breadth

    def getArea(self):
        return self.length * self.breadth

    def __str__(self):
        return f"Rectangle(length={self.length}, breadth={self.breadth})"


# derived class:Square
class Square(Rectangle):
    def __init__(self):
        super().__init__()

    def setLength(self, length):
        self.length = length
        self.breadth = length  # Ensure breadth is equal to length

    def setBreadth(self, breadth):
        # This method can be left empty or raise an exception since
        # a square's sides should always be the same
        raise ValueError("A square does not have different lengths and breadths.")

    def __str__(self):
        return f"Square(side={self.length})"


# Create a square
square = Square()

# Set the side length
square.setLength(5)

# Try setting the breadth (should raise an error)
try:
    square.setBreadth(10)
except ValueError as e:
    print(e)

# Print the area
print("Area of the square:", square.getArea())

# Print the square details
print(square)


```

```python
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model
        self.running = False

    def start(self):
        if not self.running:
            self.running = True
            print(f"{self.make} {self.model} is now running.")
        else:
            print(f"{self.make} {self.model} is already running.")

    def stop(self):
        if self.running:
            self.running = False
            print(f"{self.make} {self.model} has been stopped.")
        else:
            print(f"{self.make} {self.model} is already stopped.")

    def perform_action(self):
        # This method will be overridden by subclasses
        pass

class Car(Vehicle):
    def __init__(self, make, model):
        super().__init__(make, model)

    def perform_action(self):
        if self.running:
            print(f"{self.make} {self.model} honks!")
        else:
            print(f"{self.make} {self.model} is not running. Can't honk.")

class Motorcycle(Vehicle):
    def __init__(self, make, model):
        super().__init__(make, model)

    def perform_action(self):
        if self.running:
            print(f"{self.make} {self.model} pops a wheelie!")
        else:
            print(f"{self.make} {self.model} is not running. Can't pop a wheelie.")

# Demonstration
if __name__ == "__main__":
    # Create instances of Car and Motorcycle
    car = Car(make="Toyota", model="Camry")
    motorcycle = Motorcycle(make="Harley-Davidson", model="Sportster")

    # Start, perform action, and stop the car
    car.start()
    car.perform_action()
    car.stop()

    # Start, perform action, and stop the motorcycle
    motorcycle.start()
    motorcycle.perform_action()
    motorcycle.stop()
```
