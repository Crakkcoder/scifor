```python
# Base class for basic calculator operations
class Calculator:
    def add(self, x, y):
        return x + y

    def subtract(self, x, y):
        return x - y

    def multiply(self, x, y):
        return x * y

    def divide(self, x, y):
        if y == 0:
            return "Cannot divide by zero"
        return x / y

# Derived class with additional functionality
class ScientificCalculator(Calculator):
    def power(self, x, y):
        return x ** y

    def square_root(self, x):
        import math
        return math.sqrt(x)

# Usage example
def main():
    basic_calc = Calculator()
    print("Basic Calculator:")
    print("5 + 3 =", basic_calc.add(5, 3))
    print("5 - 3 =", basic_calc.subtract(5, 3))
    print("5 * 3 =", basic_calc.multiply(5, 3))
    print("5 / 3 =", basic_calc.divide(5, 3))

    print("\nScientific Calculator:")
    sci_calc = ScientificCalculator()
    print("2 ^ 5 =", sci_calc.power(2, 5))
    print("Square root of 25 =", sci_calc.square_root(25))
    print("5 + 3 =", sci_calc.add(5, 3))  # Inherited method

if __name__ == "__main__":
    main()
```

```python
class SumCalculator:
    def __init__(self):
        self.numbers = []

    def add_number(self, num):
        self.numbers.append(num)

    def calculate_sum(self):
        return sum(self.numbers)

class ExtendedSumCalculator(SumCalculator):
    def __init__(self):
        super().__init__()

    def average(self):
        if len(self.numbers) == 0:
            return 0
        return sum(self.numbers) / len(self.numbers)

# Example usage:
calculator = ExtendedSumCalculator()
calculator.add_number(10)
calculator.add_number(20)
calculator.add_number(30)

total_sum = calculator.calculate_sum()
avg = calculator.average()

print("Total Sum:", total_sum)
print("Average:", avg)
```
